### Importing all Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from scipy.stats import pearsonr
```

### Importing FuelConsumptionCo2 Dataset

```python
df = pd.read_csv('Data/FuelConsumptionCo2.csv')
```

### Feature Selection

```python
#Selecting some independent variables
Features = ['FUELCONSUMPTION_COMB_MPG', 'CYLINDERS', 'FUELCONSUMPTION_HWY', 'ENGINESIZE', 'FUELCONSUMPTION_COMB', 'FUELCONSUMPTION_CITY']
```

```python
#Calculating Pearson Coefficient btw independent & dependent variables
pearson_evaluation = []
for i in Features:
    coef, p = pearsonr(df[i], df['CO2EMISSIONS'])
    pearson_evaluation.append({'Feature':i, 'Correlation-Coefficient':coef, 'P-value':p})
```

```python
df_pearson = pd.DataFrame.from_records(pearson_evaluation)
```

```python
#Correlation Coefficient & P values with respect to each Feature with co2 emission
df_pearson.sort_values(by=['Correlation-Coefficient', 'P-value'], ascending=False)
```

```python
#Plotting Relation btw Independent & Dependent variables
for i in Features:
    df.plot(kind='scatter', x=i, y='CO2EMISSIONS', figsize=(10,8))
    plt.title(f'Relation Btw {i} & CO2Emission')
    plt.show()
    plt.close()
```

### Train Linear Regression

```python
#According to above analysis FUELCONSUMPTION_CITY & ENGINESIZE looks better, lets train simple linear regression
def TrainLinearRegression(X, Y):
    
    ss = StandardScaler()
    lr = LinearRegression()
    
    XNormalized = ss.fit_transform(X)
    train_x, test_x, train_y, test_y = train_test_split(XNormalized, Y, test_size=0.2, random_state=0)
    lr.fit(train_x, train_y)
    yhat_train = lr.predict(train_x)
    yhat_test = lr.predict(test_x)
    
    r2_train = lr.score(train_x, train_y)
    r2_test = lr.score(test_x, test_y)
    
    print("R-Squared (Train): ", r2_train)
    print("R-Squared (Test): ", r2_test)
    
    mse_train = mean_squared_error(train_y, yhat_train)
    mse_test = mean_squared_error(test_y, yhat_test)
    
    print("MSE (Train): ", mse_train)
    print("MSE (Test): ", mse_test)
    
    return train_x, test_x, train_y, test_y, yhat_train, yhat_test
```

### Distributed Graph

```python
def DistributedGraph(actual, predict):

    plt.figure(figsize=(12,10))

    ax1 = sns.distplot(actual, hist=False, label='Actual')
    sns.distplot(predict, hist=False, label='Predicted', ax=ax1)

    plt.legend()
    plt.title('Distributed Plot of Actual & Predicted values')
    plt.show()
    plt.close()
```

### Scatter Plot

```python
def ScatterPlot(dataset, actual, predicted):

    plt.figure(figsize=(12,10))

    plt.plot(dataset, actual, 'or', label='Actual')
    plt.plot(dataset, predicted, 'ob', label='Predicted')

    plt.legend()
    plt.title('Scatter Plot of Actual & Predicted values')
    plt.show()
    plt.close()
```

```python
#Train Simple Linear Regression Model using FUELCONSUMPTION_CITY & CO2EMISSIONS
X = df[['FUELCONSUMPTION_CITY']]
Y = df['CO2EMISSIONS']
```

```python
#Train & Evaluating the model
train_x, test_x, train_y, test_y, yhat_train, yhat_test = TrainLinearRegression(X, Y)
```

```python
#Distributed Graph btw actual & Predicted Training Values
DistributedGraph(train_y, yhat_train)
```

```python
#Distributed Graph btw actual & Predicted Testing Values
DistributedGraph(test_y, yhat_test)
```

```python
#Scatterplot on testing data
ScatterPlot(test_x, test_y, yhat_test)
```

```python
#Scatterplot on training data
ScatterPlot(train_x, train_y, yhat_train)
```
